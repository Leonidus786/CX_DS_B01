## Syntax and semantics :-


1. Syntax refers to the set of rules that define the combination of symbols that are considered to be correctly structured programs in a language. 
- In simple words syntax is about correct arrangements of words and symbols in code.
- Basically it refers to grammar of programming.

2. Semantics refers to the meaning or interpretation of the symbols, characters and commands ina language.
- It is about what your code is supposed to do when it runs.
